## Extract The Text From The PDFs ?

```python
loader = PyPDFDirectoryLoader("pdfs")

data = loader.load()

data
```

## Split The Extracted Data into Text Chunks ??

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=10)
```

```python
text_chunks = text_splitter.split_documents(data)
```

## Initializing Pinecone ??

```python
pinecone_api_key = os.environ['PINECONE_API_KEY']
```

```python
from pinecone import Pinecone

pc = Pinecone(api_key=pinecone_api_key)

index = pc.Index("test-pinecone")
```

## Create Embedding For Each of the Text Chunk ??

```python
import uuid
```

```python
# Embed text
texts = [doc.page_content for doc in text_chunks]
embeddings = embedding_model.embed_documents(texts)
```

```python
# Prepare and upload
vectors = [
    {"id": str(uuid.uuid4()), "values": vec, "metadata": {"text": text}}
    for text, vec in zip(texts, embeddings)
]
```

```python
index.upsert(vectors=vectors)
```

## Apply Similarity Search ??

```python
query = "YOLOv7 outperforms which models?"

query_vector = embedding_model.embed_query(query)
```

```python
results = index.query(
    vector=query_vector,
    top_k=5,
    include_metadata=True,
)
```

```python
print("Raw result object:")
print(results)

print("\nMatches:")
print(results.get("matches", []))
```

```python
for i, match in enumerate(results["matches"]):
    print(f"Result {i+1}")
    print(f"Score: {match['score']}")
    print(f"Text: {match['metadata']['text']}")
    print("-" * 50)
```

## Use LLM To To Wrap and Organize the Answers ?

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
from langchain.llms import HuggingFacePipeline

# 1. Load model and tokenizer
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# 2. Create a Transformers pipeline
pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer, max_length=512)

# 3. Wrap in LangChain LLM
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain_pinecone import PineconeVectorStore

# Wrap your Pinecone index
vectorstore = PineconeVectorStore(
    index_name="test-pinecone",
    embedding=embedding_model,  # HuggingFaceEmbeddings
    pinecone_api_key=pinecone_api_key
)

retriever = vectorstore.as_retriever()
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever)
```

## Q/A

```python
# 5. Ask a question
query = "YOLOv7 outperforms which models?"
answer = qa.run(query)
print(answer)
```

## Run The LLM Chat in Nice Way With sys ??

```python
import sys

while True:
    try:
        user_input = input("Input Prompt: ").strip()
        
        if user_input.lower() == 'exit':
            print("Exiting...")
            break  # or use sys.exit() if needed

        if not user_input:
            continue  # Skip empty input

        result = qa({"query": user_input})
        print(f"Answer: {result.get('result', 'No answer returned')}")

    except KeyboardInterrupt:
        print("\nInterrupted by user. Exiting...")
        break

    except Exception as e:
        print(f"Error: {e}")
```
